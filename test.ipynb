# Test for the program

We use three locations to test whether this project works well. The first two locations are 'Paris Las Vegas' hotel and 'The D Las Vegas' hotel, and we successfully find restaurants around them. We then test the "Nanjing University" in China, because we cannot find restaurants in Las Vegas arond the location , it outputs 'Oops, there is no such restaurant near you :('.

```python
import json
import requests
import Visualization #The moduel we code for map visualization
from Dinner_Function import dinner #The moduel we code for restaurants selection

def what_for_dinner(person_site,formatted_address,radius=500):
    
    #Import clusters data
    with open('clusters.json','rb') as clusters_init:
        clusters_data = json.loads(clusters_init.read())
        
    #import clusters center
    with open('clusters_centers.json','rb') as clusters_coordinates:
        clusters_coordinates = json.loads(clusters_coordinates.read())
        
    results = dinner(person_site,clusters_coordinates, clusters_data,radius)
    return Visualization.ResultsVisualization(results,formatted_address)   
```

## 'Paris Las Vegas' Hotel, distance is 1000 meters

<li>Get the formatted location of you

```python
address = 'Paris Las Vegas'.replace(' ','+')+'+Las+Vegas'
response = requests.get('https://maps.googleapis.com/maps/api/geocode/json?address=%s&key=AIzaSyBmpmNHzWutfb5CY7K3PeZZbvU4KAld3rM'%address)
resp_json_payload = response.json()

formatted_address = resp_json_payload['results'][0]['formatted_address']
print('The formatted address of your current location:',formatted_address)
person_site = [resp_json_payload['results'][0]['geometry']['location']['lat'],
              resp_json_payload['results'][0]['geometry']['location']['lng']]
radius = 1000
```

<li>Get the restaurants around you

```python
what_for_dinner(person_site,formatted_address,radius=float(radius))
```
