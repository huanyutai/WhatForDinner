<h2>Input Data</h2>
<li>list of cluster vertexs
<li>list of dicts: restaurants name and location(lat,lon)
<li>current location(lat,lon)

```python
with open('clusters.json') as fp:
    data = fp.read()
data
```

```python
import json
result_list = json.loads(data)
result_list
```

<h2>Divide sectors based on clusters</h2>
Using K-NN algorithm to label the restaurants in to different clusters

<h2>Mark restaurants on the map

```python
feature_list = []
feature_collection = {'type':"FeatureCollection","features":feature_list}
color = ['#FF3333','#FFCC00','#9933FF','#339900','#3366FF']
i = 0
for results in result_list:
    col = color[i]
    i = i + 1
    for restaurant in results:
        name = restaurant['name']
        lat = restaurant['latitude']
        lon = restaurant['longitude']
        resta_dict = {'type':"Feature",
                      "geometry":{"type":"Point", "coordinates":[lon, lat]},
                      "properties":{"name":name,'marker-color':col}}
        feature_list.append(resta_dict)
```

```python
import geojsonio
import json
geojsonio.display(json.dumps(feature_collection))
```
