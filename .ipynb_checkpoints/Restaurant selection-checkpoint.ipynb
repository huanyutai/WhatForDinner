```python
import urllib
from bs4 import BeautifulSoup
import requests
```

```python
import json
with open('clusters_final.json','r') as json_data:
    data = json_data.read()
restaurants = json.loads(data)[1:]
restaurants[0]
```

```python
Names = []
Locations = []
for restaurant in restaurants:
    if restaurant['name'] != None and restaurant['address']!=None:
        name = restaurant['name']
        location = restaurant['address']+', Las Vegas'
        Names.append(name)
        Locations.append(location)
```

```python
Name = Names[0]
Location = Locations[0]
Name_url = urllib.request.quote(Name)
Location_url = urllib.request.quote(Location)
url = 'https://www.yelp.com/search?find_desc='+Name_url+'&find_loc='+Location_url
headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.90 Safari/537.36'}
response = requests.get(url,headers=headers,verify=False)
```

```python
n = len(Names)
Links = []
for i in range(n):
    Name = Names[i]
    Location = Locations[i]
    Name_url = urllib.request.quote(Name)
    Location_url = urllib.request.quote(Location)
    url = 'https://www.yelp.com/search?find_desc='+Name_url+'&find_loc='+Location_url
    headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.90 Safari/537.36'}
    response = requests.get(url,headers=headers,verify=False)
    results_page = BeautifulSoup(response.content,'lxml')
    search_results = results_page.find_all('a',{'class':'biz-name js-analytics-click'})
    for search_result in search_results:
        if search_result.find('span').get_text() == Name:
            url_1 = 'https://www.yelp.com'+search_result.get('href')
            print(url_1)
            Links.append(url_1)   
            break
```
